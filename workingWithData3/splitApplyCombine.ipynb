```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
# dataset --> split by index --> apply something to vals to get new vals --> combine
dframe_wine = pd.read_csv("winequality-red.csv", sep = ";")
dframe_wine.head()
```

```python
# rank wines by highest alcohol content
def ranker(df):
    df["alc_content_rank"] = np.arange(len(df)) + 1
    return df
```

```python
dframe_wine.sort_values("alcohol", ascending = False, inplace = True)
```

```python
dframe_wine.reset_index(drop = True, inplace = True) # avoiding ambiguity
dframe_wine = dframe_wine.groupby("quality").apply(ranker, include_groups = True)
```
