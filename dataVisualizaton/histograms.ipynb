```python
import numpy as np
import pandas as pd
from numpy.random import randn
# stats
from scipy import stats
# plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
dataset1 = randn(100)
```

```python
plt.hist(dataset1)
plt.show()
```

```python
dataset2 = randn(80)
plt.hist(dataset2, color = "indianred")
plt.show()
```

```python
# density is normalizing since the two data are different sizes, alpha(value 0 to 1) is transparency, bins default is 10
plt.hist(dataset1, density = True, color = "indianred", alpha = 0.5, bins = 20)
plt.hist(dataset2, density = True, alpha = 0.5, bins = 20)
plt.show()
```

```python
data1 = randn(1000)
data2 = randn(1000)
```

```python
sns.jointplot(x = data1, y = data2)
plt.show()
```

```python
# hex bins for more concise viz
sns.jointplot(x = data1, y = data2, kind = "hex")
plt.show()
```
