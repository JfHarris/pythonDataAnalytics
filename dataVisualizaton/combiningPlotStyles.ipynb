```python
import numpy as np
from numpy.random import randn
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = randn(100)
```

```python
# distplot deprecated, adding kde = True to displot will replicate it
sns.displot(dataset, bins = 25, kde = True) # default bins = 10
plt.show()
```

```python
# displot without histogram
sns.displot(dataset, rug = True, kind = "kde")
plt.show()
```

```python
# controlling specific plots within a plot
sns.displot(dataset, bins = 25, kde = True, legend = True,
           line_kws = {"linestyle": "dashed"})
plt.show()
```

```python
axisHist = sns.histplot(dataset, kde = False, stat = "density", color = "blue")
sns.kdeplot(dataset, color = "red")
plt.show()
```
