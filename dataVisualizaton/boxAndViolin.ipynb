```python
import numpy as np
from numpy.random import randn
import pandas as pd
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dataset1 = randn(100)
dataset2 = randn(100)
```

```python
sns.boxplot([dataset1,dataset2])
plt.show()
```

```python
# include whiskers
sns.boxplot([dataset1,dataset2], whis = np.inf)
plt.show()
```

```python
sns.boxplot(dataset1, vert = False)
plt.show()
```

```python
# normal distribution
dataset1 = stats.norm(0, 5).rvs(100)

# 2 gamma distributions concatenated together
dataset2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                          stats.gamma(5).rvs(50) * -1])
```

```python
sns.boxplot([dataset1,dataset2], whis = np.inf)
plt.show()
```

```python
sns.violinplot([dataset1,dataset2])
plt.show()
```

```python
sns.violinplot(dataset2, bw_method = 0.01)
plt.show()
```

```python
sns.violinplot(dataset1)
plt.show()
```

```python
sns.violinplot(dataset1, inner = "stick")
plt.show()
```
