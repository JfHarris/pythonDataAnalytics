```python
import numpy as np
from numpy.random import randn
import pandas as pd
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.lmplot(data = tips, x = "total_bill", y = "tip")
plt.show()
```

```python
sns.lmplot(data = tips, x = "total_bill", y = "tip",
          scatter_kws = {"marker": "o", "color": "indianred"},
          line_kws = {"linewidth": 1, "color": "blue"})
plt.show()
```

```python
sns.lmplot(data = tips, x = "total_bill", y = "tip", order = 4,
          scatter_kws = {"marker": "o", "color": "indianred"},
          line_kws = {"linewidth": 1, "color": "blue"})
plt.show()
```

```python
# adding tip pct column to tips
tips["tip_pct"] = 100 * tips["tip"] / tips["total_bill"]
tips.head()
```

```python
sns.lmplot(data = tips, x = "size", y = "tip_pct")
plt.show()
```

```python
sns.lmplot(data = tips, x = "size", y = "tip_pct", x_jitter = 0.1)
plt.show()
```

```python
sns.lmplot(data = tips, x = "size", y = "tip_pct", x_estimator = np.mean)
plt.show()
```

```python
sns.lmplot(data = tips, x = "total_bill", y = "tip_pct", hue = "sex", markers = ["x", "o"])
plt.show()
```

```python
sns.lmplot(data = tips, x = "total_bill", y = "tip_pct", hue = "day")
plt.show()
```

```python
# local regression
sns.lmplot(data = tips, x = "total_bill", y = "tip_pct", lowess = True,
            line_kws = {"color": "black"})
plt.show()
```

```python
# lmplot uses regplot(regression plot)
sns.regplot(data = tips, x = "total_bill", y = "tip_pct")
plt.show()
```

```python
fig, (axis1, axis2) = plt.subplots(1,2, sharey = True)
sns.regplot(data = tips, x = "total_bill", y = "tip_pct", ax = axis1)
sns.violinplot(y = tips["tip_pct"], x = tips["size"], color = "red", ax = axis2)
plt.show()
```
